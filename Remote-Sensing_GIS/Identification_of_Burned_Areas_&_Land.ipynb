## Forest and Land Fire Prone Areas of Ogan Komering Ilir

```python
# Read shapefile into a GeoDataFrame
gdf_RK = gpd.read_file("/content/drive/MyDrive/PYTHON_GEEMAP/BC-LOGISMAP-August-2024/Data-Day2/Data-SHP-Day2/Rawan_Karhutla.shp")
gdf_RK.head()
```

```python
fig, ax3 = plt.subplots(figsize=(14,8))
gdf_RK.plot(ax=ax3,column='Kelas',legend=True,cmap="Accent",
         legend_kwds={'loc':'best','bbox_to_anchor': (1, 1)})  # Color based on a column
ax3.set_title("Sebaran Area Rawan Karhutla\nOgan Komering Ilir Tahun 2019")
plt.show()
```

```python
plt.bar(gdf_RK.Kelas,gdf_RK.Shape_Area,color='green')
plt.ylabel("Rawan Karhutla Ogan Komering Ilir")
plt.show()
```

```python
# Change Shape_Area units from square meters to hectares
gdf_RK['Shape_Area_Ha'] = gdf_RK['Shape_Area'] / 10000

plt.bar(gdf_RK.Kelas, gdf_RK.Shape_Area_Ha, color='green')
plt.ylabel("Rawan Karhutla Ogan Komering Ilir (Hektar)")
plt.show()
```

```python
# Filtering GeoDataFrame for 'Class' with description 'Low, Medium, High, Very High'
gdf_RK_rendah = gdf_RK[gdf_RK['Kelas'] == 'Rendah']
gdf_RK_sedang = gdf_RK[gdf_RK['Kelas'] == 'Sedang']
gdf_RK_tinggi = gdf_RK[gdf_RK['Kelas'] == 'Tinggi']
gdf_RK_sangat_tinggi = gdf_RK[gdf_RK['Kelas'] == 'Sangat Tinggi']

ha = 10000

# Sum the 'Shape_Area' values ​​from the filtered GeoDataFrame
total_area_rendah = gdf_RK_rendah['Shape_Area'].sum() / 10000
total_area_sedang = gdf_RK_sedang['Shape_Area'].sum() / 10000
total_area_tinggi = gdf_RK_tinggi['Shape_Area'].sum() / 10000
total_area_sangat_tinggi = gdf_RK_sangat_tinggi['Shape_Area'].sum() / 10000

print("Area prone to forest and land fires with the description 'Low':", total_area_rendah, 'Ha')
print("Area prone to forest and land fires with the description 'Moderate':", total_area_sedang, 'Ha')
print("Area prone to forest and land fires with the description 'High':", total_area_tinggi, 'Ha')
print("The area prone to forest and land fires is classified as 'Very High':", total_area_sangat_tinggi, 'Ha')
```

```python
# Data for pie chart
labels = ['Rendah', 'Sedang', 'Tinggi', 'Sangat Tinggi']
sizes = [total_area_rendah, total_area_sedang, total_area_tinggi, total_area_sangat_tinggi]
colors = ['maroon', 'red', 'orange', 'yellow']

# Creating a pie chart
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# Adding a title
ax.set_title('Percentage of Area Prone to Forest and Land Fires in Ogan Komering Ilir in 2019')

# Displaying a pie chart
plt.show()
```

```python
# Creating a pie chart
fig, ax = plt.subplots()
wedges, texts, autotexts = ax.pie(sizes, labels=labels, colors=colors, autopct='', startangle=90)

# Add area description
for i, autotext in enumerate(autotexts):
    autotext.set_text(f"{sizes[i]:.2f}")

# Add title
ax.set_title('Comparison of Areas Prone to Forest and Land Fires in Ogan Komering Ilir in 2019')

# Displays pie chart
plt.show()
```

## Identification of the Affected Area

```python
# Make sure both GeoDataFrames have the same CRS
gdf2 = gdf.to_crs(gdf_AT.crs)
# Lakukan overlay spasial (spatial overlay)
overlay_result = gpd.overlay(gdf2, gdf_AT, how='intersection')

# Make sure both GeoDataFrames have the same CRS
gdf_TL2 = gdf_TL.to_crs(gdf_AT.crs)
# Perform spatial overlay
overlay_result_TL = gpd.overlay(gdf_TL2, gdf_AT, how='intersection')
```

### Kecamatan Terdampak

```python
# Show the overlay results of the affected districts
fig, ax6 = plt.subplots(figsize=(8, 8))
overlay_result.plot(ax=ax6, column='KECAMATAN', legend=True, cmap='tab20',
                    legend_kwds={'loc': 'best', 'bbox_to_anchor': (1.4, 1)})
ax6.set_title("Kecamatan Terdampak")
plt.show()
```

```python
#  Hitung luas area terbakar per kecamatan
luas_per_kecamatan = overlay_result.groupby('KECAMATAN')['geometry'].apply(lambda geom: geom.area.sum())

# Print the results
luas_per_kecamatan
```

```python
# Plot of the burned area in each sub-district
luas_per_kecamatan_sorted = luas_per_kecamatan.sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(8, 6))
luas_per_kecamatan_sorted.plot(kind='bar', color='blue')
plt.title('Burned Area per District in Ogan Komering Ilir in 2019')
plt.xlabel('Kecamatan')
plt.ylabel('Luas Area Terbakar (Ha)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Calculate the burned area per sub-district in hectares
luas_per_kecamatan_ha = overlay_result.groupby('KECAMATAN')['geometry'].apply(lambda geom: geom.area.sum() / 10000)

# Cetak hasilnya
print(luas_per_kecamatan_ha)

# Sort data by area burned
luas_per_kecamatan_ha_sorted = luas_per_kecamatan_ha.sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(10, 6))
luas_per_kecamatan_ha_sorted.plot(kind='bar', color='blue')
plt.title('Burned Area per District (Hectares) in Ogan Komering Ilir in 2019')
plt.xlabel('Kecamatan')
plt.ylabel('Luas Area Terbakar (Hektar)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Show the results of the Affected Land Cover overlay
fig, ax7 = plt.subplots(figsize=(8, 8))
overlay_result_TL.plot(ax=ax7, column='Keterangan', legend=True, cmap='jet',
                    legend_kwds={'loc': 'best', 'bbox_to_anchor': (1.55, 1)})
ax7.set_title("Tutupan Lahan Terdampak")
plt.show()
```

### Affected Land Cover

```python
# Calculate the burned area per land cover
luas_per_TL = overlay_result_TL.groupby('Keterangan')['geometry'].apply(lambda geom: geom.area.sum())

# Print the results
luas_per_TL
```

```python
# Sort data by area burned
luas_per_TL_sorted = luas_per_TL.sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(8, 6))
luas_per_TL_sorted.plot(kind='bar', color='purple')
plt.title('Burned Area per Land Cover in Ogan Komering Ilir in 2019')
plt.xlabel('Tutupan Lahan')
plt.ylabel('Luas Area Terbakar (Ha)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Calculate the burned area per land cover in hectares
luas_per_TL_ha = overlay_result_TL.groupby('Keterangan')['geometry'].apply(lambda geom: geom.area.sum() / 10000)

# Print the results
print(luas_per_TL_ha)

# Sort data by area burned
luas_per_TL_ha_sorted = luas_per_TL_ha.sort_values(ascending=False)

# Create a bar chart
plt.figure(figsize=(8, 6))
luas_per_TL_ha_sorted.plot(kind='bar', color='purple')
plt.title('Burned Area per Land Cover (Hectare) in Ogan Komering Ilir in 2019')
plt.xlabel('Tutupan Lahan')
plt.ylabel('Luas Area Terbakar (Hektar)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Calculate the total area burned
total_area_terbakar = luas_per_TL_ha.sum()

# Calculate the percentage of burned area per land cover
persentase_luas_per_TL = (luas_per_TL_ha / total_area_terbakar) * 100

# Sort data by percentage of burned area
persentase_luas_per_TL_sorted = persentase_luas_per_TL.sort_values(ascending=False)

# Create a pie chart
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(persentase_luas_per_TL_sorted,
                                  #labels=persentase_luas_per_TL_sorted.index,
                                  autopct='', startangle=90,
                                  textprops={'fontsize': 10},
                                  colors=['#800000','#982B1C','#CD5C08','#FFAD60','#180161',
                                          '#4F1787','#021526','#03346E','#6EACDA','#694F8E',
                                          '#B692C2','#1A3636','#40534C','#677D6A','#399918'])

# Arrange the label text so that it does not overlap.
plt.setp(autotexts, size=8, weight="bold")
#plt.setp(texts, size=10)

# Add a separate legend with percentages
legend_labels = [f"{label} ({percentage:.1f}%)"
                 for label, percentage in zip(persentase_luas_per_TL_sorted.index, persentase_luas_per_TL_sorted)]
plt.legend(wedges, legend_labels,
          title="Tutupan Lahan",
          loc='center left',
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Percentage of Burned Area per Land Cover (Hectare) in Ogan Komering Ilir in 2019')
plt.show()
```
