```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv('Global_LandTemp_SeaSurfTemp.csv')
ds.head(10)
```

```python
ds_MAM = np.array(ds.MAM)
ds_time = np.array(ds.Year)
```

```python
plt.plot(ds_time,ds_MAM)
plt.plot(ds_time,np.array(ds.JJA))
plt.plot(ds_time,np.array(ds.SON))
plt.show()
```
